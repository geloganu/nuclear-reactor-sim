```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
file='output.out'
backsc,transm,absorb,therma=np.loadtxt(file,skiprows=1,unpack=True)
d=np.arange(1,61)
```

```python
fig = plt.subplots(figsize=(6,4))
fig = plt.plot(d,backsc,label='Backscattered')
fig = plt.plot(d,transm,label='Transmitted')
fig = plt.plot(d,absorb,label='Absorbed')
fig = plt.plot(d,therma,label='Thermalized')

plt.legend()
plt.xlabel('Graphite Slab Thickness (cm)')
plt.ylabel('% of Emitted Neutrons')

plt.savefig('neutron_percentage.png')
```

```python
exit_energies='exit_energies.out'
unmod='unmod_energies.out'

ene=np.loadtxt(exit_energies,unpack=True)
unmod=np.loadtxt(unmod,unpack=True)

ex_count,bin=np.histogram(ene,bins=100)
un_count,bin=np.histogram(unmod,bins=100)

bin=bin[0:100]
```

```python
plt.subplots(figsize=(6,4))
plt.plot(bin,ex_count,label='Exit Energy Spectrum')
plt.plot(bin,un_count,label='Unmoderated Energy Spectrum')

plt.legend()
plt.xlabel('Energy (MeV)')
plt.ylabel('Count')

plt.savefig('exit_energy.png')
```
